```python
class DivIcon(MacroElement):
    def __init__(self, html='', size=(30,30), anchor=(0,0), style=''):
        """TODO : docstring here"""
        super(DivIcon, self).__init__()
        self._name = 'DivIcon'
        self.size = size
        self.anchor = anchor
        self.html = html
        self.style = style

        self._template = Template(u"""
            {% macro header(this, kwargs) %}
              <style>
                .{{this.get_name()}} {
                    {{this.style}}
                    }
              </style>
            {% endmacro %}
            {% macro script(this, kwargs) %}
                var {{this.get_name()}} = L.divIcon({
                    className: '{{this.get_name()}}',
                    iconSize: [{{ this.size[0] }},{{ this.size[1] }}],
                    iconAnchor: [{{ this.anchor[0] }},{{ this.anchor[1] }}],
                    html : "{{this.html}}",
                    });
                {{this._parent.get_name()}}.setIcon({{this.get_name()}});
            {% endmacro %}
            """)
```

```python
import folium
import pandas as pd
from folium import plugins
import warnings; warnings.simplefilter('ignore')

map_hooray = folium.Map(location=[30.355709, 112.747444],
                    zoom_start = 5) 

df_acc = pd.read_csv('lat_lon_inputs.csv')

# Ensure you're handing it floats
df_acc['Latitude'] = df_acc['Latitude'].astype(float)
df_acc['Longitude'] = df_acc['Longitude'].astype(float)

# Filter the DF for rows, then columns, then remove NaNs
heat_df = df_acc[['Latitude', 'Longitude']]

# Create weight column, using date
heat_df['Weight'] = df_acc['Year']
heat_df['Weight'] = heat_df['Weight'].astype(float)
heat_df = heat_df.dropna(axis=0, subset=['Latitude','Longitude', 'Weight'])

# List comprehension to make out list of lists
heat_data = [[[row['Latitude'],row['Longitude']] for index, 
              row in heat_df[heat_df['Weight'] == i].iterrows()] for i in range(1938,1948)]

# Plot it on the map
hm = plugins.HeatMapWithTime(heat_data,auto_play=False,max_opacity=0.8)
hm.add_to(map_hooray)


################################################## Add Points ##################################################
folium.map.Marker(
    [31.225549, 121.472240]).add_to(map_hooray) # Shanghai
folium.map.Marker(
    [39.198227, 117.347550]).add_to(map_hooray) # Tianjin
folium.map.Marker(
    [22.407772, 114.121550]).add_to(map_hooray) # Hong Kong
folium.map.Marker(
    [25.045294, 102.732739]).add_to(map_hooray) # Kunming


################################################## Add Points ##################################################
folium.map.Marker(
    [31.090433, 120.939331],
    icon=DivIcon(
#         size=(150,36),
#         anchor=(0,0),
        html='Shanghai',
        style="""
            font-size:20px;
            background-color: transparent;
            border-color: transparent;
            text-align: center;
            """
        )
    ).add_to(map_hooray)
folium.map.Marker(
    [39.198227, 117.347550],
    icon=DivIcon(
#         size=(150,36),
#         anchor=(0,0),
        html='Tianjin',
        style="""
            font-size:20px;
            background-color: transparent;
            border-color: transparent;
            text-align: center;
            """
        )
    ).add_to(map_hooray)
folium.map.Marker(
    [22.407772, 114.121550],
    icon=DivIcon(
#         size=(150,36),
#         anchor=(0,0),
        html='Hong&nbspKong',
        style="""
            font-size:20px;
            background-color: transparent;
            border-color: transparent;
            text-align: center;
            """
        )
    ).add_to(map_hooray)
folium.map.Marker(
    [25.045294, 102.732739],
    icon=DivIcon(
#         size=(150,36),
#         anchor=(0,0),
        html='Kunming',
        style="""
            font-size:20px;
            background-color: transparent;
            border-color: transparent;
            text-align: center;
            """
        )
    ).add_to(map_hooray)

folium.map.Marker(
    [26.563404, 129.080642],
    icon=DivIcon(
        size=(150,30),
        anchor=(150,0),
        html='<bold>Fatalities&nbspfrom Terrorist&nbspActivity </br>(1938&nbspto&nbsp1947</bold>)',
        style="""
            font-size:36px;
            background-color: transparent;
            border-color: transparent;
            text-align: center;
            """
        )
    ).add_to(map_hooray)
# Display the map

map_hooray
```
